```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset("AITC-Traffic-Density/Traffic-Object-Detection")

print(dataset)
```

```python
import os

image_dir = "traffic_images"
os.makedirs(image_dir, exist_ok=True)

image_paths = []

for i, item in enumerate(dataset["train"]):
    img = item["image"]
    path = os.path.join(image_dir, f"img_{i}.jpg")
    img.save(path)
    image_paths.append(path)

print(f"Saved {len(image_paths)} images")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # small + fast baseline
```

```python
def detect_and_count(model, image_path, conf=0.25):
    results = model.predict(
        source=image_path,
        device="cpu",   # IMPORTANT: baseline = CPU
        conf=conf,
        verbose=False
    )

    result = results[0]

    car_count = 0
    pedestrian_count = 0

    for cls_id in result.boxes.cls.tolist():
        cls_id = int(cls_id)

        if cls_id == 0:      # person
            pedestrian_count += 1
        elif cls_id == 2:    # car
            car_count += 1

    return car_count, pedestrian_count
```

```python
import time

timings = []
results_summary = []

# Warm-up (important for fair timing)
for i in range(3):
    _ = model.predict(source=image_paths[i], device="cpu", verbose=False)

start_total = time.perf_counter()

for path in image_paths:
    start = time.perf_counter()

    cars, people = detect_and_count(model, path)

    end = time.perf_counter()

    latency = end - start
    timings.append(latency)

    results_summary.append({
        "image": path,
        "cars": cars,
        "pedestrians": people,
        "latency_ms": latency * 1000
    })

end_total = time.perf_counter()

total_time = end_total - start_total
avg_latency = sum(timings) / len(timings)
throughput = len(image_paths) / total_time

print("=== BASELINE RESULTS (CPU, batch=1) ===")
print(f"Images:        {len(image_paths)}")
print(f"Total time:    {total_time:.4f} s")
print(f"Avg latency:   {avg_latency*1000:.2f} ms")
print(f"Throughput:    {throughput:.2f} images/sec")
```

```python
import time
import cv2
import numpy as np
import pandas as pd

def benchmark_breakdown(model, image_paths, conf=0.25, num_images=50):
    """
    Measures time spent in:
      - preprocessing
      - inference
      - postprocessing
    """

    results = []

    # limit number of images (optional, keeps runtime reasonable)
    image_subset = image_paths[:num_images]

    for img_path in image_subset:
        # --- Preprocessing ---
        t0 = time.perf_counter()

        img = cv2.imread(img_path)
        img_resized = cv2.resize(img, (640, 640))  # typical YOLO input size
        img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)

        t1 = time.perf_counter()

        # --- Inference ---
        results_model = model.predict(
            source=img_rgb,
            device="cpu",   # change to "cuda" if you want GPU breakdown
            conf=conf,
            verbose=False
        )

        t2 = time.perf_counter()

        # --- Postprocessing ---
        result = results_model[0]

        car_count = 0
        pedestrian_count = 0

        if result.boxes is not None and result.boxes.cls is not None:
            for cls_id in result.boxes.cls.tolist():
                cls_id = int(cls_id)
                if cls_id == 0:
                    pedestrian_count += 1
                elif cls_id == 2:
                    car_count += 1

        t3 = time.perf_counter()

        results.append({
            "preprocessing_ms": (t1 - t0) * 1000,
            "inference_ms": (t2 - t1) * 1000,
            "postprocessing_ms": (t3 - t2) * 1000
        })

    df = pd.DataFrame(results)

    print("=== Breakdown (average per image) ===")
    print(f"Preprocessing:  {df['preprocessing_ms'].mean():.2f} ms")
    print(f"Inference:      {df['inference_ms'].mean():.2f} ms")
    print(f"Postprocessing: {df['postprocessing_ms'].mean():.2f} ms")

    return df


# Run it
breakdown_df = benchmark_breakdown(
    model=model,
    image_paths=image_paths,
    conf=0.25,
    num_images=50
)

breakdown_df.head()
```

```python
import matplotlib.pyplot as plt

avg_pre = breakdown_df["preprocessing_ms"].mean()
avg_inf = breakdown_df["inference_ms"].mean()
avg_post = breakdown_df["postprocessing_ms"].mean()

plt.figure(figsize=(6, 6))
plt.bar(["Preprocessing", "Inference", "Postprocessing"],
        [avg_pre, avg_inf, avg_post])

plt.ylabel("Time (ms)")
plt.title("Execution Time Breakdown")
plt.show()
```
